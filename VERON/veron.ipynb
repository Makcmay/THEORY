```python
import openpyxl

# Открытие файла res.xlsx
wb = openpyxl.load_workbook('res.xlsx')

# Получение листа
ws = wb.active

# Определение столбца руководителя (например, столбец A)
manager_col = 'B'

# Получение уникальных значений из столбца руководителей
managers = set([cell.value for cell in ws[manager_col][1:]])

# Создание нового файла для каждого руководителя и копирование соответствующих строк
for manager in managers:
    # Создание нового файла с названием "<имя руководителя>.xlsx"
    new_wb = openpyxl.Workbook()
    new_ws = new_wb.active

    # Копирование строки заголовка с сохранением цветов ячеек
    header_row = ws[1]
    for i, cell in enumerate(header_row):
        new_ws.cell(row=1, column=i+1, value=cell.value)
        if cell.fill.start_color.index != 'FFFFFFFF':
            new_ws.cell(row=1, column=i+1).fill = cell.fill

    # Копирование строк данных руководителя в новый файл
    row_idx = 2  # начинаем со второй строки (первая строка - заголовок)
    for row in ws.iter_rows(min_row=row_idx, values_only=True):
        if row[0] == manager:
            new_ws.append(row)

    # Сохранение нового файла
    new_wb.save(f'{manager}.xlsx')
```

```python
import pandas as pd
import openpyxl
from openpyxl import Workbook
from openpyxl.styles import Color, PatternFill
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.cell import cell

# чтение файла csv
# df = pd.read_csv("result.csv")
df = pd.read_excel('result.xlsx', sheet_name='Лист1')
# определение уникальных руководителей
managers = df['Руководитель'].unique()

# создание новых файлов Excel для каждого руководителя
for manager in managers:
    wb = Workbook()
    ws = wb.active

    # выбор строк, соответствующих текущему руководителю
    current_manager = df[df['Руководитель'] == manager]

    # запись данных в файл Excel с сохранением цветового выделения
    for r in dataframe_to_rows(current_manager, index=False, header=True):
        ws.append(r)
        for cell in ws:
            if cell.fill.start_color.index != '00000000':
                cell.fill = PatternFill(start_color=cell.fill.start_color.index,
                                        end_color=cell.fill.end_color.index,
                                        fill_type=cell.fill.fill_type)
    # сохранение файла Excel
    wb.save(f"{manager}.xlsx")
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import PatternFill

# Читаем данные из файла input.xlsx
df = pd.read_excel('result.xlsx', sheet_name='Лист1')

# Группируем данные по руководителям
groups = df.groupby('Руководитель')

# Проходимся по группам и сохраняем данные каждой группы в отдельный файл
for name, group in groups:
    # Создаем новый Excel файл
    wb = Workbook()

    # Получаем активный лист
    ws = wb.active

    # Копируем данные из группы в лист
    for r in dataframe_to_rows(group, index=False, header=True):
        ws.append(r)

    # Сохраняем цвет ячеек в столбце "Итог"
    for row in ws.iter_rows(min_row=1, min_col=1):
        for cell in row:
            if cell.column == 'Итог':
                fill = PatternFill(fill_type='solid', start_color=cell.fill.fgColor.rgb)
                cell.fill = fill

    # Сохраняем файл с именем равным имени руководителя
    wb.save(f'{name}.xlsx')
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils import get_column_letter
from openpyxl.styles import Color, PatternFill

# Считываем данные из файла excel
df = pd.read_excel('result.xlsx')

# Группируем данные по уникальным значениям признака "руководитель"
groups = df.groupby('Руководитель')

# Проходимся по каждой группе и сохраняем данные в отдельный файл
for name, group in groups:
    # Создаем новый файл
    wb = Workbook()
    ws = wb.active

    # Копируем заголовки столбцов в новый файл
    for col_idx, column in enumerate(group.columns):
        col_letter = get_column_letter(col_idx + 1)
        ws[f"{col_letter}1"] = column

    # Копируем данные из группы в новый файл
    for row_idx, row_data in group.iterrows():
        for col_idx, cell_data in enumerate(row_data):
            col_letter = get_column_letter(col_idx + 1)
            cell_address = f"{col_letter}{row_idx + 2}"
            ws[cell_address] = cell_data

            # Сохраняем цвет ячейки
            original_cell = df.iloc[row_idx, col_idx]
            if isinstance(original_cell, str):
                fill = PatternFill(start_color=original_cell, end_color=original_cell, fill_type='solid')
                ws[cell_address].fill = fill

    # Сохраняем новый файл с названием "руководитель.xlsx"
    wb.save(f"{name}.xlsx")
```
