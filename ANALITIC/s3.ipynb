```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', sep = ',', encoding='latin-1')
df.tail(5)
```

1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

1.4 Удалите признаки Memory и ScreenResolution


```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(" ")[0]) #сплит применяется к одной строке, когда строк много нужен аплай
df.head()
```

Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def convert(x):
    memory = x.split(' ')[0] # делим на список по первому значению
    if memory.endswith('GB'): # endswith проверяет окончание строки ГБ или нет, возвращает тру или фалс
        res = memory.replace('GB', '') #если заканчивается на ГБ то меняем на пустую строку
    elif memory.endswith('TB'): #если заканчивается на ТБ
        res = float(memory.replace('TB', '')) * 1024 # меняем на пустую, преобразовывает во флот и умножаем на 1024
    return int(res) #результат в инте

df['Memory_Amount_GB'] = df['Memory'].apply(convert) 


# df.drop(columns=['Memory_Amount'], inplace=True)

df.head()
```

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
def convert2(x):
    memory = x.split(" ")#  сплитим строку из мемори по пробелу
    res = memory[1:] # берем все данные кроме первых
    return ' '.join(res) # соединяем
    # return ' '.join(x.split(' ')[1: ]) аналогичная запись

df['Memory_Type'] = df['Memory'].apply(convert2)
```

```python
df.head()
```

1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```

2.2 Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD

2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный


```python
df['SSD'] = 0
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1 #если действительно ссд то = 1 f.loc[df['Memory_Type'](маска поиска) == 'SSD', 'SSD'(столбец)](на что поменять)
df
```

Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный


```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float64') 
df
```
